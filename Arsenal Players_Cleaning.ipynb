```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\USERR\Documents\Arsenal Players python web scrapping project csv\Arsenal Players.csv")
df
```

```python
df.info()
```

```python
df.head()
```

```python
## df[] = df[] to save what we have cleaned as pandas doesnt automatically save, for players without market value python shows - hence we have to remove
## it, .str.strip) removes invisible white spaces from the beginig and end of a string, error = coerce converts what it cant make numeric into nan
df['market_value'] = df['market_value'].str.replace('€', '')
df['market_value'] = df['market_value'].str.replace('m', '')
df['market_value'] = df['market_value'].str.replace('-', '')
df['market_value'] = df['market_value'].str.strip()
df['market_value'] = pd.to_numeric(df['market_value'], errors = 'coerce')
df
```

```python
## to rename the column to signify millions
df = df.rename(columns = {'market_value' : 'market_value_millons'})
df
```

```python
## to remove , amd m from height
df['height'] = df['height'].str.replace('m', '')
df['height'] = df['height'].str.replace(',', '')
df
```

```python
print(df['height'].head())
```

```python
## lets rename the column
df = df.rename(columns = {'height' : 'height_cm'})
df
```

```python
## now to extract the age from the dob, find an opening bracket\( then (d+) find one or more digits, then \) find a closing bracket
df['age'] = df['dob'].str.extract(r'\((\d+)\)')
df['age'].head()
```

```python
df
```

```python
print(df['dob'].head())
```

```python
## to convert age to integer
df['age'] = df['age'].astype(int)
```

```python
## to emove the extracted age from dob column, \s* remove white spaces befoe the bracket, \( find open bracket, .* match anything inside the bracket
## \) find closing bracket regex = true tell python that ths a regular expression not a literal string so python isnt looking for \s*\(.*\)
df['dob'] = df['dob'].str.replace(r'\s*\(.*\)', '', regex=True).str.strip()
```

```python
## to convert dob to proper date
df['dob'] = pd.to_datetime(df['dob'], format='%d/%m/%Y')
print(df['dob'].head())
```

```python
df['joined'] = pd.to_datetime(df['joined'], format='%d/%m/%Y')
```

```python
df['contract'] = pd.to_datetime(df['contract'], format='%d/%m/%Y')
df
```
