```python
## we are telling it to find a table tag that has the class items it is crucial to put the table 
## as it ensures it returns only rows in that table
table = soup.find('table', class_="items")
print(table)
```

```python
## we collecct the players info and store in a dictionary 
player = {
    "jersey_number": cells[0].text.strip(),
    "name": cells[3].text.strip(),
    "position": cells[4].text.strip(),
    "nationality": cells[6].find('img')['title'],
    "dob": cells[5].text.strip(),
    "height": cells[7].text.strip(),
    "foot": cells[8].text.strip(),
    "joined": cells[9].text.strip(),
    "contract": cells[11].text.strip(),
    "market_value": cells[12].text.strip()
}

print(player)
```

```python
## we create an empty list and create a for loop to extract all player info into a dictionary theb we append it intothe list
players = []  # empty list to collect all players

for row in rows:
    cells = row.find_all('td')
    
    if len(cells) < 13:  # skip rows that aren't player rows
        continue
        
    player = {
        "jersey_number": cells[0].text.strip(),
        "name": cells[3].text.strip(),
        "position": cells[4].text.strip(),
        "nationality": cells[6].find('img')['title'] if cells[6].find('img') else None,
        "dob": cells[5].text.strip(),
        "height": cells[7].text.strip(),
        "foot": cells[8].text.strip(),
        "joined": cells[9].text.strip(),
        "contract": cells[11].text.strip(),
        "market_value": cells[12].text.strip()
    }
    
    players.append(player)

print(len(players))
```

```python
## now lest save it into the data frame
## and save it to a csv file create the folder and copy the path name it and end with csv

import pandas as pd
df = pd.DataFrame(players)
df.to_csv(r"C:\Users\USERR\Documents\Arsenal Players python web scrapping project csv\Arsenal Players.csv", index = False)
df
```
